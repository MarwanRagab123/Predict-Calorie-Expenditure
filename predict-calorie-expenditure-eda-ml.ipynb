# Visualize the data

```python
sns.scatterplot(data=train_df,
x="Age",
y="Calories",
hue="Sex")
plt.show()
```

```python
hue_color={
    "male":"blue",
    "female":"red"
}
sns.countplot(x="Sex",data=train_df,palette=hue_color)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


g = sns.relplot(x="Duration",
                y="Calories",
                data=train_df,
                kind="scatter",
                col="Sex",
                )


g.set(ylim=(0, train_df['Calories'].max() * 1.1))  # 10% padding on top
g.set(xlim=(0, train_df['Duration'].max() * 1.1))  # 10% padding on right

plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


g = sns.relplot(x="Heart_Rate",
                y="Calories",
                data=train_df,
                kind="scatter",
                hue="Sex",
                style="Sex")


g.set(ylim=(0, train_df['Calories'].max() * 1.1))  # 10% padding on top
g.set(xlim=(0, train_df['Heart_Rate'].max() * 1.1))  # 10% padding on right

plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


g = sns.relplot(x="Body_Temp",
                y="Calories",
                data=train_df,
                kind="line",
                )


g.set(ylim=(0, train_df['Calories'].max() * 1.1))  # 10% padding on top
g.set(xlim=(0, train_df['Body_Temp'].max() * 1.1))  # 10% padding on right

plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```

```python
plt.scatter(train_df['Calories'],train_df['Duration'])
plt.xlabel('Calories')
plt.ylabel('Duration')
plt.title('Calories vs Duration')
plt.show()
```

```python
plt.scatter(train_df['Calories'],train_df['Heart_Rate'])
plt.xlabel('Calories')
plt.ylabel('Heart_Rate')
plt.title('Calories vs Heart_Rate')
plt.show()
```

```python
train_df['Sex']=train_df['Sex'].map({'male':1,'female':0})
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(train_df.corr(), annot=True, cmap='coolwarm', ax=ax)
ax.set_title('Correlation Matrix')
plt.show()
```

```python
cols = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']
for col in cols:
    plt.figure(figsize=(8, 4))
    sns.catplot(x=train_df['Sex'],y=train_df[col],kind='box',palette={'male':'blue','female':'red'})
    plt.title(f'Boxplot of {col}')
    plt.show()
```
